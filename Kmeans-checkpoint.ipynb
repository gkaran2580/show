```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
```

```python
X,y=make_blobs(n_samples=300,centers=6,cluster_std=0.80,random_state=0)  # std means gaps or random state means data is fxed forever
#print(X)
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=6)
km.fit(X)
pred=km.predict(X)
print(pred)
```

```python
plt.scatter(X[:,0],X[:,1],c=pred,cmap='winter')
centers=km.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',s=300,alpha=0.5)
```

#how kmeans algo works
STEP1: choose the no of k clusters
step2:select at random k points(the centroid)
step3:assign each data points to closet centroid
step4:compute and place the new centroid of each cluster
step5:reassign each datapoint to the new closet centroid
    if any Reassignmnet took place,go to step4
    otherwise model is ready

```python
from sklearn.cluster import KMeans
import numpy as np
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img=Image.open('bird.jpg')
img_np=np.asarray(img)
img_np.shape
```

```python
pixels=img_np.reshape(img_np.shape[0]*img_np.shape[1],img_np.shape[2])
pixels.shape
```

```python
model=KMeans(n_clusters=16)
model.fit(pixels)
```

```python
pixels_centroids=model.labels_
cluster_centers=model.cluster_centers_
pixels_centroids
```

```python
cluster_centers
```

```python
final=np.zeros((pixels_centroids.shape[0],3))
```

```python
final.shape
```

```python
for cluster_no in range(16):
    final[pixels_centroids==cluster_no]=cluster_centers[cluster_no]
final
```

```python
# reshape
comp_image=final.reshape(img_np.shape[0],img_np.shape[1],3)
comp_image.shape
```

```python
#saving compressed image
comp_image=Image.fromarray(np.uint8(comp_image))
comp_image.save('bird1.jpg')
```

```python
fname='bird.jpg'
fname1='bird1.jpg'
img1=mpimg.imread(fname,0)
img2=mpimg.imread(fname1,0)
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(20,20))
ax1.imshow(img1)
ax1.set_title('Orginal Image')
ax2.imshow(img2)
ax2.set_title('Compressed Image')
plt.show()
```
